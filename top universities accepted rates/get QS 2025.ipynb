```python
import requests
import pandas as pd

# URL of the JSON data source
data_url = 'https://www.qschina.cn/sites/default/files/qs-rankings-data/cn/2057712_indicators.txt'

# Headers to mimic a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.3945.88 Safari/537.36',
    'Referer': 'https://www.qschina.cn/university-rankings/world-university-rankings/2025'
}

# Fetch the data
response = requests.get(data_url, headers=headers)
if response.status_code != 200:
    print(f"Failed to fetch data: HTTP {response.status_code}")
else:
    # Parse the JSON content
    data = response.json()

# Correctly extract data from the JSON structure
records = []
for entry in data.get('data', []):  # Iterate over the main 'data' key
    record = {
        'Rank': entry.get('overall_rank', '').strip(),
        'University': entry.get('uni', '').strip(),
        'Country': entry.get('location', '').strip(),
        'Overall Score': entry.get('score', '').strip(),
        'Academic Reputation': entry.get('ar_score', '').strip(),
        'Employer Reputation': entry.get('er_score', '').strip(),
        'Faculty/Student Ratio': entry.get('fsr_score', '').strip(),
        'Citations per Faculty': entry.get('cpf_score', '').strip(),
        'International Faculty Ratio': entry.get('ifr_score', '').strip(),
        'International Student Ratio': entry.get('isr_score', '').strip()
    }
    records.append(record)

# Create a DataFrame
if records:
    df = pd.DataFrame(records)
    # Save to CSV
    #df.to_csv('qs_world_university_rankings_2025.csv', index=False, encoding='utf-8')
    #print("Data successfully saved to 'qs_world_university_rankings_2025.csv'.")
else:
    print("No data found in the response.")
```
