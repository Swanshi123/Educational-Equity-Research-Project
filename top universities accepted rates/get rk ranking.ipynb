```python
import requests
import re
```

```python
url = 'https://www.shanghairanking.cn/_nuxt/static/1734427793/rankings/bcur/2024/payload.js'
response = requests.get(url=url)
response.encoding = 'utf-8'
html = response.text
html
```

```python
import pandas as pd
# 使用正则表达式匹配大学中文名和英文名
univNameCn_pattern = r'univNameCn:"(.*?)"'
univNameEn_pattern = r'univNameEn:"([^"]*)"'

univNameCn_matches = re.findall(univNameCn_pattern, html)
univNameEn_matches = re.findall(univNameEn_pattern, html)

from itertools import zip_longest
chinese_names = [name if name != '' else pd.NA for name in univNameCn_matches]
english_names = [name if name != '' else pd.NA for name in univNameEn_matches]
combined = list(zip_longest(chinese_names, english_names, fillvalue=pd.NA))

# 创建DataFrame
df = pd.DataFrame(combined, columns=['Chinese Name', 'English Name'])
df
```
