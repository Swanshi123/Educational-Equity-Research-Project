```python
def assign_value(row):
    if row['211 project universities'] in df_new['rktop100'].values:
        return 2
    elif row['211 project universities'] in df_new['qstop71'].values:
        return 3
    else:
        return 1  # or any value to denote it does not match
# Apply the function to the DataFrame
df_new['Result'] = df_new.apply(assign_value, axis=1)
df_new
```
